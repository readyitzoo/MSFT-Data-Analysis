```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, pacf
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
from pmdarima.arima import auto_arima

# Load data
df = pd.read_csv("MSFT.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'] >= '2018-01-01']
df.sort_values('Date', inplace=True)
series = df['Close']

# Plot original series
plt.plot(df['Date'], series)
plt.title("Microsoft Daily Closing Prices")
plt.show()

# Stationarity test
result = adfuller(series.dropna())
print("ADF Statistic:", result[0])
print("p-value:", result[1])

# Make series stationary if needed
if result[1] > 0.05:
    series_diff = series.diff().dropna()
else:
    series_diff = series.copy()

# PACF to estimate order
pacf_vals = pacf(series_diff, nlags=20)
plt.stem(range(len(pacf_vals)), pacf_vals)
plt.title("PACF Plot")
plt.show()

# Manual AR(2) model
model = AutoReg(series_diff, lags=2, old_names=False).fit()
forecast = model.predict(start=len(series_diff), end=len(series_diff)+199)

# Auto ARIMA model
auto_model = auto_arima(series, seasonal=False, suppress_warnings=True)
auto_forecast = auto_model.predict(n_periods=200)

# Forecast comparison
future_index = pd.date_range(start=df['Date'].iloc[-1], periods=201, freq='D')[1:]
plt.plot(df['Date'], series, label="Original")
plt.plot(future_index, forecast, label="Manual AR(2) Forecast")
plt.plot(future_index, auto_forecast, label="Auto ARIMA Forecast")
plt.legend()
plt.title("Forecast Comparison")
plt.show()

# Evaluation
print("Manual AR(2) Coefficients:", model.params)
print("Auto ARIMA Order:", auto_model.order)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.stattools import adfuller

# === STEP 1: Load and Prepare the Dataset ===

# Load the MSFT stock data (make sure MSFT.csv is in the same directory)
df = pd.read_csv("MSFT.csv")

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter the data to include only rows from 2018 onwards
df = df[df['Date'] >= '2018-01-01']

# Sort the data by date (just to ensure chronological order)
df.sort_values('Date', inplace=True)

# Extract the closing prices as a NumPy array
series_raw = df['Close'].values

# === STEP 2: Stationarity Check and Differencing if Needed ===

# Perform Augmented Dickey-Fuller test to check for stationarity
adf_result = adfuller(series_raw)

# If p-value > 0.05, the series is not stationary, so we difference it
if adf_result[1] > 0.05:
    series = np.diff(series_raw)  # First difference
    differenced = True            # Flag to revert differencing later
else:
    series = series_raw.copy()
    differenced = False

# === STEP 3: Define Sample Autocovariance Function ===

# Computes the sample autocovariance at a given lag
def sample_autocovariance(x, lag):
    x_mean = np.mean(x)
    return np.sum((x[:len(x)-lag] - x_mean) * (x[lag:] - x_mean)) / len(x)

# === STEP 4: Durbin-Levinson Algorithm Implementation ===

# Estimates AR coefficients  for AR(p) using recursive Durbin-Levinson algorithm
def durbin_levinson(x, p_max):
    gamma = [sample_autocovariance(x, lag) for lag in range(p_max+1)]  # Sample autocovariances
    phi = np.zeros((p_max+1, p_max+1))  # Coefficient matrix [k][j]
    sigma2 = np.zeros(p_max+1)         # Prediction error variance

    # Initial coefficient for AR(1)
    phi[1,1] = gamma[1] / gamma[0]
    sigma2[1] = gamma[0] * (1 - phi[1,1]**2)

    # Recursive computation for AR(2) to AR(p_max)
    for k in range(2, p_max+1):
        sum_phi_gamma = sum(phi[k-1,j]*gamma[k-j] for j in range(1, k))
        phi[k,k] = (gamma[k] - sum_phi_gamma) / sigma2[k-1]
        for j in range(1, k):
            phi[k,j] = phi[k-1,j] - phi[k,k]*phi[k-1,k-j]
        sigma2[k] = sigma2[k-1]*(1 - phi[k,k]**2)

    return phi

# === STEP 5: Estimate AR(2) Coefficients ===

phi_matrix = durbin_levinson(series, p_max=2)
phi_1, phi_2 = phi_matrix[2,1], phi_matrix[2,2]  # Extract  and 

# === STEP 6: Manual Forecasting with AR(2) Model ===

history = list(series)  # Initialize with original data
forecast = []

# Generate 200 forecasted values
for _ in range(200):
    next_val = phi_1 * history[-1] + phi_2 * history[-2]
    forecast.append(next_val)
    history.append(next_val)

# === STEP 7: Revert Differencing (if applied) ===

# If we differenced the data earlier, we need to revert it to original scale
if differenced:
    last_value = df['Close'].iloc[-1]
    forecast = np.cumsum(forecast) + last_value

# === STEP 8: Create Future Date Index for Plotting ===

last_date = df['Date'].iloc[-1]
forecast_dates = pd.date_range(start=last_date, periods=201, freq='D')[1:]

# === STEP 9: Plot Original Series and Forecast ===

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Original Series')
plt.plot(forecast_dates, forecast, label='Manual AR(2) Forecast')
plt.legend()
plt.title('Microsoft Daily Closing Prices with Manual AR(2) Forecast')

# Show only the year on the x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.grid(True)
plt.show()
```

```python
#!/usr/bin/env python3
"""
Time Series Analysis Project - Complete Implementation
Student: [Your Name]
Date: May 2025

This project implements:
1. ARIMA modeling of Microsoft stock prices with manual parameter selection
2. Durbin-Levinson algorithm implementation and comparison with PACF
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.linalg import toeplitz
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# PART 1: TIME SERIES MODELING CLASSES
# =============================================================================

class TimeSeriesAnalyzer:
    def __init__(self, data):
        self.data = data
        self.original_data = data.copy()
        
    def plot_series(self, title="Time Series"):
        """Plot the time series"""
        plt.figure(figsize=(12, 6))
        plt.plot(self.data.index, self.data.values)
        plt.title(title)
        plt.xlabel('Date')
        plt.ylabel('Close Price')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    def check_stationarity(self, data=None):
        """Augmented Dickey-Fuller test for stationarity"""
        if data is None:
            data = self.data
        
        def adf_test(timeseries):
            y = timeseries.values
            n = len(y)
            dy = np.diff(y)
            y_lag = y[:-1]
            
            X = np.column_stack([np.ones(n-1), y_lag, np.arange(1, n)])
            
            try:
                beta = np.linalg.lstsq(X, dy, rcond=None)[0]
                residuals = dy - X @ beta
                mse = np.sum(residuals**2) / (n - 4)
                
                se_beta = np.sqrt(mse * np.linalg.inv(X.T @ X)[1, 1])
                t_stat = beta[1] / se_beta
                
                return {'statistic': t_stat, 'pvalue': 1 if t_stat > -2.86 else 0.01}
            except:
                return {'statistic': 0, 'pvalue': 1}
        
        result = adf_test(data)
        print(f"ADF Statistic: {result['statistic']:.4f}")
        print(f"p-value: {result['pvalue']:.4f}")
        
        if result['pvalue'] <= 0.05:
            print("Series is stationary")
            return True
        else:
            print("Series is non-stationary")
            return False
    
    def difference_series(self, d=1):
        """Apply differencing to make series stationary"""
        for i in range(d):
            self.data = self.data.diff().dropna()
        return self.data
    
    def acf_pacf_manual(self, lags=40):
        """Manual implementation of ACF and PACF"""
        data = self.data.dropna()
        n = len(data)
        mean = data.mean()
        
        # ACF calculation
        acf_vals = []
        for k in range(lags + 1):
            if k == 0:
                acf_vals.append(1.0)
            else:
                c_k = np.sum((data[k:] - mean) * (data[:-k] - mean)) / n
                c_0 = np.sum((data - mean) ** 2) / n
                acf_vals.append(c_k / c_0)
        
        # PACF calculation using Yule-Walker equations
        pacf_vals = [1.0]
        
        for k in range(1, lags + 1):
            if k == 1:
                pacf_vals.append(acf_vals[1])
            else:
                R = np.array([[acf_vals[abs(i-j)] for j in range(k)] for i in range(k)])
                r = np.array([acf_vals[i] for i in range(1, k+1)])
                try:
                    phi = np.linalg.solve(R, r)
                    pacf_vals.append(phi[-1])
                except:
                    pacf_vals.append(0)
        
        return np.array(acf_vals), np.array(pacf_vals)
    
    def plot_acf_pacf(self, lags=40):
        """Plot ACF and PACF"""
        acf_vals, pacf_vals = self.acf_pacf_manual(lags)
        
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
        
        # Plot ACF
        ax1.stem(range(len(acf_vals)), acf_vals)
        ax1.set_title('Autocorrelation Function (ACF)')
        ax1.set_xlabel('Lag')
        ax1.set_ylabel('ACF')
        ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        ax1.axhline(y=1.96/np.sqrt(len(self.data)), color='red', linestyle='--', alpha=0.5)
        ax1.axhline(y=-1.96/np.sqrt(len(self.data)), color='red', linestyle='--', alpha=0.5)
        ax1.grid(True)
        
        # Plot PACF
        ax2.stem(range(len(pacf_vals)), pacf_vals)
        ax2.set_title('Partial Autocorrelation Function (PACF)')
        ax2.set_xlabel('Lag')
        ax2.set_ylabel('PACF')
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        ax2.axhline(y=1.96/np.sqrt(len(self.data)), color='red', linestyle='--', alpha=0.5)
        ax2.axhline(y=-1.96/np.sqrt(len(self.data)), color='red', linestyle='--', alpha=0.5)
        ax2.grid(True)
        
        plt.tight_layout()
        plt.show()
        
        return acf_vals, pacf_vals

class ARIMAModel:
    def __init__(self, data):
        self.data = data
        self.params = None
        self.residuals = None
        
    def fit_ar(self, p):
        """Fit AR(p) model using Yule-Walker equations"""
        data = self.data.dropna()
        n = len(data)
        
        # Calculate autocorrelations
        acf_vals = []
        mean = data.mean()
        c_0 = np.sum((data - mean) ** 2) / n
        
        for k in range(p + 1):
            if k == 0:
                acf_vals.append(1.0)
            else:
                c_k = np.sum((data[k:] - mean) * (data[:-k] - mean)) / n
                acf_vals.append(c_k / c_0)
        
        # Solve Yule-Walker equations
        R = np.array([[acf_vals[abs(i-j)] for j in range(p)] for i in range(p)])
        r = np.array([acf_vals[i] for i in range(1, p+1)])
        
        try:
            phi = np.linalg.solve(R, r)
            self.params = {'ar': phi, 'const': mean * (1 - np.sum(phi))}
            
            self.calculate_residuals_ar(p)
            return phi
        except:
            print("Error fitting AR model")
            return None
    
    def calculate_residuals_ar(self, p):
        """Calculate residuals for AR model"""
        data = self.data.dropna()
        n = len(data)
        residuals = np.zeros(n)
        
        for t in range(p, n):
            pred = self.params['const']
            for i in range(p):
                pred += self.params['ar'][i] * data.iloc[t-1-i]
            residuals[t] = data.iloc[t] - pred
        
        self.residuals = residuals[p:]
    
    def forecast_ar(self, steps=20, p=None):
        """Forecast using AR model"""
        if self.params is None:
            print("Model not fitted")
            return None
        
        data = self.data.dropna()
        forecasts = []
        
        last_obs = data.tail(len(self.params['ar'])).values[::-1]
        
        for _ in range(steps):
            forecast = self.params['const']
            for i, coef in enumerate(self.params['ar']):
                forecast += coef * last_obs[i]
            
            forecasts.append(forecast)
            last_obs = np.roll(last_obs, 1)
            last_obs[0] = forecast
        
        return np.array(forecasts)

# =============================================================================
# PART 2: DURBIN-LEVINSON ALGORITHM
# =============================================================================

class DurbinLevinsonAnalyzer:
    def __init__(self):
        pass
    
    def durbin_levinson(self, acf_values, max_order):
        """Implement Durbin-Levinson algorithm"""
        phi_mm = np.zeros(max_order + 1)
        phi_mm[0] = 1.0
        
        if max_order == 0:
            return phi_mm
        
        phi_mm[1] = acf_values[1] / acf_values[0]
        v = acf_values[0] * (1 - phi_mm[1]**2)
        
        phi = np.zeros((max_order + 1, max_order + 1))
        phi[1, 1] = phi_mm[1]
        
        for m in range(2, max_order + 1):
            numerator = acf_values[m]
            for k in range(1, m):
                numerator -= phi[m-1, k] * acf_values[m-k]
            
            phi_mm[m] = numerator / v
            
            for k in range(1, m):
                phi[m, k] = phi[m-1, k] - phi_mm[m] * phi[m-1, m-k]
            
            phi[m, m] = phi_mm[m]
            v = v * (1 - phi_mm[m]**2)
        
        return phi_mm
    
    def generate_ar2_data(self, n=200, phi1=1.0, phi2=-0.9, sigma=1.0):
        """Generate AR(2) process"""
        if abs(phi2) >= 1 or phi1 + phi2 >= 1 or phi2 - phi1 >= 1:
            print("Warning: AR(2) parameters may not satisfy stationarity conditions")
        
        np.random.seed(42)
        epsilon = np.random.normal(0, sigma, n + 100)
        
        x = np.zeros(n + 100)
        x[0] = epsilon[0]
        x[1] = phi1 * x[0] + epsilon[1]
        
        for t in range(2, n + 100):
            x[t] = phi1 * x[t-1] + phi2 * x[t-2] + epsilon[t]
        
        return x[-n:]
    
    def estimate_acf(self, data, max_lag):
        """Estimate sample autocorrelation function"""
        n = len(data)
        mean = np.mean(data)
        
        acf = np.zeros(max_lag + 1)
        acf[0] = np.sum((data - mean)**2) / n
        
        for h in range(1, max_lag + 1):
            acf[h] = np.sum((data[h:] - mean) * (data[:-h] - mean)) / n
        
        acf = acf / acf[0]
        return acf
    
    def compare_with_builtin_pacf(self, data, max_order=10):
        """Compare D-L results with built-in PACF function"""
        acf_vals = self.estimate_acf(data, max_order)
        dl_pacf = self.durbin_levinson(acf_vals, max_order)
        
        # Simplified PACF calculation
        def simple_pacf(data, nlags):
            n = len(data)
            mean = np.mean(data)
            
            acf = np.zeros(nlags + 1)
            for k in range(nlags + 1):
                if k == 0:
                    acf[k] = 1.0
                else:
                    c_k = np.sum((data[k:] - mean) * (data[:-k] - mean)) / n
                    c_0 = np.sum((data - mean) ** 2) / n
                    acf[k] = c_k / c_0
            
            pacf = np.zeros(nlags + 1)
            pacf[0] = 1.0
            
            for k in range(1, nlags + 1):
                if k == 1:
                    pacf[k] = acf[1]
                else:
                    R = np.array([[acf[abs(i-j)] for j in range(k)] for i in range(k)])
                    r = np.array([acf[i] for i in range(1, k+1)])
                    try:
                        phi = np.linalg.solve(R, r)
                        pacf[k] = phi[-1]
                    except:
                        pacf[k] = 0
            
            return pacf
        
        builtin_pacf = simple_pacf(data, max_order)
        return dl_pacf, builtin_pacf

# =============================================================================
# MSFT DATA ANALYSIS FUNCTIONS
# =============================================================================

def analyze_msft_data():
    """Main function to analyze MSFT stock data"""
    
    print("="*60)
    print("MICROSOFT STOCK TIME SERIES ANALYSIS")
    print("="*60)
    
    # Load MSFT data
    try:
        df = pd.read_csv('MSFT.csv')
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        
        # Use last 2000 observations for manageable analysis
        ts_data = df['Close'].tail(2000)
        
        print(f"Data loaded: {len(ts_data)} observations")
        print(f"Period: {ts_data.index[0].strftime('%Y-%m-%d')} to {ts_data.index[-1].strftime('%Y-%m-%d')}")
        print(f"Price range: ${ts_data.min():.2f} to ${ts_data.max():.2f}")
        
    except FileNotFoundError:
        print("MSFT.csv not found. Creating sample data for demonstration...")
        # Create sample data that mimics stock behavior
        np.random.seed(42)
        dates = pd.date_range('2020-01-01', periods=2000, freq='D')
        returns = np.random.normal(0.0005, 0.02, 2000)
        prices = 100 * np.exp(np.cumsum(returns))
        ts_data = pd.Series(prices, index=dates)
        print("Using sample data for demonstration")
    
    # Initialize analyzer
    analyzer = TimeSeriesAnalyzer(ts_data)
    
    print("\n1. ORIGINAL TIME SERIES ANALYSIS")
    analyzer.plot_series("Microsoft Stock Prices (MSFT)")
    
    print("\n2. STATIONARITY TEST")
    is_stationary = analyzer.check_stationarity()
    
    d = 0  # Track differencing order
    if not is_stationary:
        print("\n3. APPLYING FIRST DIFFERENCING")
        analyzer.difference_series(d=1)
        d = 1
        analyzer.plot_series("MSFT Returns (First Differenced)")
        
        print("Testing stationarity after first differencing:")
        is_stationary_diff = analyzer.check_stationarity()
        
        if not is_stationary_diff:
            print("Applying second differencing...")
            analyzer.difference_series(d=1)
            d = 2
            analyzer.plot_series("Second Differenced MSFT")
    
    print(f"\nFinal differencing order (d): {d}")
    
    print("\n4. ACF AND PACF ANALYSIS")
    acf_vals, pacf_vals = analyzer.plot_acf_pacf(lags=25)
    
    # Analyze significant lags
    n = len(analyzer.data)
    significance_level = 1.96 / np.sqrt(n)
    
    print(f"\nSignificance threshold: {significance_level:.4f}")
    
    significant_pacf = [(i, pacf_vals[i]) for i in range(1, min(6, len(pacf_vals))) 
                       if abs(pacf_vals[i]) > significance_level]
    significant_acf = [(i, acf_vals[i]) for i in range(1, min(6, len(acf_vals))) 
                      if abs(acf_vals[i]) > significance_level]
    
    print("Significant PACF lags:", significant_pacf)
    print("Significant ACF lags:", significant_acf)
    
    # Model selection based on patterns
    if significant_pacf:
        suggested_p = max([lag for lag, _ in significant_pacf])
        print(f"\nSuggested AR order (p): {suggested_p}")
    else:
        suggested_p = 1
        print(f"\nDefaulting to AR(1) model")
    
    print("\n5. MODEL FITTING")
    model = ARIMAModel(analyzer.data)
    
    # Fit AR model
    print(f"Fitting AR({suggested_p}) model...")
    ar_params = model.fit_ar(p=suggested_p)
    
    if ar_params is not None:
        print(f"AR({suggested_p}) parameters:")
        for i, param in enumerate(ar_params):
            print(f"  {i+1} = {param:.6f}")
        print(f"  Constant = {model.params['const']:.6f}")
        
        # Check model adequacy with residuals
        if model.residuals is not None:
            residual_mean = np.mean(model.residuals)
            residual_std = np.std(model.residuals)
            print(f"\nResidual diagnostics:")
            print(f"  Mean: {residual_mean:.6f} (should be close to 0)")
            print(f"  Std Dev: {residual_std:.6f}")
    
    print("\n6. FORECASTING")
    forecasts = model.forecast_ar(steps=20, p=suggested_p)
    
    if forecasts is not None:
        print("20-step ahead forecasts:")
        for i, forecast in enumerate(forecasts[:10], 1):
            print(f"  Step {i:2d}: {forecast:.6f}")
        if len(forecasts) > 10:
            print("  ...")
            for i, forecast in enumerate(forecasts[10:], 11):
                print(f"  Step {i:2d}: {forecast:.6f}")
    
    # Plot results
    plot_forecast_results(analyzer, forecasts, suggested_p)
    
    return analyzer, model, forecasts, suggested_p, d

def plot_forecast_results(analyzer, forecasts, p):
    """Plot forecasting results"""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Plot 1: Recent actual data
    recent_data = analyzer.data.tail(100)
    axes[0,0].plot(recent_data.index, recent_data.values, 'b-', linewidth=2)
    axes[0,0].set_title('Recent Data (Last 100 observations)')
    axes[0,0].grid(True)
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Plot 2: Forecasts
    axes[0,1].plot(range(1, len(forecasts)+1), forecasts, 'r--o', linewidth=2, markersize=4)
    axes[0,1].set_title(f'20-Step Ahead Forecasts (AR({p}))')
    axes[0,1].set_xlabel('Forecast Step')
    axes[0,1].grid(True)
    
    # Plot 3: Combined view (last 20 actual + 20 forecasts)
    last_20 = analyzer.data.tail(20)
    combined_x = list(range(-19, 1)) + list(range(1, 21))
    combined_y = list(last_20.values) + list(forecasts)
    
    axes[1,0].plot(range(-19, 1), last_20.values, 'b-', linewidth=2, label='Actual')
    axes[1,0].plot(range(1, 21), forecasts, 'r--', linewidth=2, label='Forecast')
    axes[1,0].axvline(x=0, color='black', linestyle=':', alpha=0.7)
    axes[1,0].set_title('Actual vs Forecasted Values')
    axes[1,0].set_xlabel('Time (0 = forecast start)')
    axes[1,0].legend()
    axes[1,0].grid(True)
    
    # Plot 4: Forecast statistics
    forecast_mean = np.mean(forecasts)
    forecast_std = np.std(forecasts)
    axes[1,1].hist(forecasts, bins=8, alpha=0.7, color='red', edgecolor='black')
    axes[1,1].axvline(forecast_mean, color='blue', linestyle='--', 
                     label=f'Mean: {forecast_mean:.4f}')
    axes[1,1].set_title('Forecast Distribution')
    axes[1,1].set_xlabel('Forecast Value')
    axes[1,1].set_ylabel('Frequency')
    axes[1,1].legend()
    axes[1,1].grid(True)
    
    plt.tight_layout()
    plt.show()

def run_durbin_levinson_analysis():
    """Run Durbin-Levinson algorithm analysis"""
    
    print("\n" + "="*60)
    print("DURBIN-LEVINSON ALGORITHM ANALYSIS")
    print("="*60)
    
    dl_analyzer = DurbinLevinsonAnalyzer()
    
    print("1. GENERATING AR(2) DATA")
    print("Parameters:  = 1.0,  = -0.9")
    ar2_data = dl_analyzer.generate_ar2_data(n=200, phi1=1.0, phi2=-0.9)
    
    # Plot generated data
    plt.figure(figsize=(12, 4))
    plt.plot(ar2_data, 'b-', linewidth=1)
    plt.title('Generated AR(2) Process: X = 1.0X - 0.9X + ')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()
    
    print(f"Generated {len(ar2_data)} observations")
    print(f"Mean: {np.mean(ar2_data):.4f}")
    print(f"Std Dev: {np.std(ar2_data):.4f}")
    
    print("\n2. APPLYING DURBIN-LEVINSON ALGORITHM")
    max_order = 10
    dl_pacf, builtin_pacf = dl_analyzer.compare_with_builtin_pacf(ar2_data, max_order)
    
    print("\n3. COMPARISON RESULTS")
    print("Order\tD-L PACF\tBuilt-in PACF\tDifference")
    print("-" * 50)
    for i in range(min(8, len(dl_pacf))):
        diff = abs(dl_pacf[i] - builtin_pacf[i])
        print(f"{i:2d}\t{dl_pacf[i]:8.6f}\t{builtin_pacf[i]:8.6f}\t{diff:8.6f}")
    
    # Plot comparison
    plt.figure(figsize=(12, 6))
    orders = range(len(dl_pacf))
    plt.plot(orders, dl_pacf, 'o-', label='Durbin-Levinson', markersize=6, linewidth=2)
    plt.plot(orders, builtin_pacf, 's-', label='Built-in PACF', markersize=4, linewidth=2)
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    # Significance bounds
    cutoff = 1.96 / np.sqrt(200)
    plt.axhline(y=cutoff, color='red', linestyle='--', alpha=0.5, label=f'95% CI ({cutoff:.3f})')
    plt.axhline(y=-cutoff, color='red', linestyle='--', alpha=0.5)
    
    plt.xlabel('Order')
    plt.ylabel('PACF')
    plt.title('Comparison: Durbin-Levinson vs Built-in PACF')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print("\n4. MODEL ORDER IDENTIFICATION")
    significant_orders = []
    for i in range(1, len(dl_pacf)):
        if abs(dl_pacf[i]) > cutoff:
            significant_orders.append((i, dl_pacf[i]))
    
    print("Significant PACF values:")
    for order, value in significant_orders:
        print(f"  Order {order}: {value:.6f}")
    
    if len(significant_orders) >= 2:
        print(f"\nIdentified model: AR({significant_orders[-1][0]}) - matches true AR(2)!")
    elif len(significant_orders) == 1:
        print(f"\nIdentified model: AR({significant_orders[0][0]})")
    else:
        print("\nNo significant PACF values detected")
    
    return dl_pacf, builtin_pacf, ar2_data

def main():
    """Main execution function"""
    print("TIME SERIES ANALYSIS PROJECT")
    print("Student: [Your Name]")
    print("="*60)
    
    # Part 1: MSFT Analysis
    try:
        analyzer, model, forecasts, p, d = analyze_msft_data()
        print(f"\nPart 1 Summary:")
        print(f"- Model: ARIMA({p},{d},0)")
        print(f"- Forecasts generated: {len(forecasts) if forecasts is not None else 0}")
    except Exception as e:
        print(f"Error in Part 1: {e}")
    
    # Part 2: Durbin-Levinson Analysis
    try:
        dl_pacf, builtin_pacf, ar2_data = run_durbin_levinson_analysis()
        
        # Calculate implementation accuracy
        max_diff = np.max(np.abs(dl_pacf - builtin_pacf))
        mean_diff = np.mean(np.abs(dl_pacf - builtin_pacf))
        
        print(f"\nPart 2 Summary:")
        print(f"- Maximum difference: {max_diff:.8f}")
        print(f"- Mean absolute difference: {mean_diff:.8f}")
        print(f"- Implementation accuracy: {100*(1-mean_diff):.2f}%")
        
    except Exception as e:
        print(f"Error in Part 2: {e}")
    
    print("\n" + "="*60)
    print("ANALYSIS COMPLETE")
    print("="*60)
    print("Next steps:")
    print("1. Review all plots and numerical results")
    print("2. Write your report based on the analysis")
    print("3. Include model justification and interpretations")
    print("4. Compare your manual results with auto_arima if available")

if __name__ == "__main__":
    main()
```
